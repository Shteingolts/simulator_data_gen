```python
import network
import matplotlib.pyplot as plt
import os

sim_dir = "/home/sergey/python/simulator_data_gen/network_data"
```

```python
os.chdir(sim_dir)
networks = [network.Network.from_data_file((os.path.join(d, "network.lmp"))) for d in os.listdir(sim_dir) if d.isdigit()]
networks
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2)

y = [net.coordination_number for net in networks]
n_nodes = [len(net.atoms) for net in networks]
n_edges = [len(net.bonds) for net in networks]
# bond_coeffs = [net.bonds[0].bond_coefficient for net in networks]
axs[0][0].hist(y)
axs[0][1].hist(n_nodes)
axs[1][0].hist(n_edges)
# axs[1][1].hist(bond_coeffs)
plt.show()
```
